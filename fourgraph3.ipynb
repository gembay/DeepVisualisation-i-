```python
#Which continents have the highest average GDP per country?
```

```python
x =ds.continent
y =ds.gdp
```

```python
gdp_by_continent = ds.groupby('continent')['gdp'].mean()
gdp_by_continent.plot(kind="bar")
plt.xlabel("continents")
plt.ylabel("GDP")
plt.title("Which continents have the highest average GDP per country?")
plt.show()
```

```python
#How do birth rates compare between rural and urban populations across regions?
```

```python
birthrates_by_continent = ds.groupby('region')['birth_rate'].mean()
birthrates_by_continent.plot(kind="bar")
plt.ylabel("birthrate")
plt.xlabel("region")
plt.title("How do birth rates compare between rural and urban populations across regions?")
plt.show()
```

```python
#How does life expectancy vary across democracy types?
```

```python
birthrates_by_continent = ds.groupby('democracy_type')['life_expectancy'].mean()
birthrates_by_continent.plot(kind="bar")
plt.ylabel("life_expectancy")
plt.xlabel("democracy_type")
plt.title("#How does life expectancy vary across democracy types?")
plt.show()
```

```python
#What is the distribution of life expectancy across all countries?
```

```python
plt.figure(figsize=(15,8))
plt.hist(ds["life_expectancy"],bins=20)
plt.xlabel("life_expectancy")
plt.title("#How does life expectancy vary across democracy types?")
plt.show()
```

```python
#How is GDP distributed across countries?
```

```python
plt.figure(figsize=(15,8))
plt.hist(ds["gdp"],bins=20)
plt.xlabel("gdp")
plt.title("How is GDP distributed across countries?")
plt.show()
```

```python
#What is the distribution of fertility rates across continents?
```

```python
fertility = ds.groupby('continent')['fertility_rate']
plt.figure(figsize=(15,8))
fertility.hist(bins=20)
plt.xlabel("continent")
plt.ylabel("fertility_rate")
plt.title("What is the distribution of fertility rates across continents")
plt.show()
```

```python

# 4. Composition: Stacked bar plot of electricity production sources for top 10 GDP countries
import numpy as np
top_10_gdp = ds.nlargest(10, 'gdp')
sources = [
    'electricty_production_coal_pct', 'electricty_production_hydroelectric_pct',
    'electricty_production_nuclear_pct', 'electricty_production_gas_pct',
    'electricty_production_oil_pct', 'electricty_production_renewable_pct'
]
labels = ['Coal', 'Hydroelectric', 'Nuclear', 'Gas', 'Oil', 'Renewable']

plt.figure(figsize=(12, 8))
bottom = np.zeros(len(top_10_gdp))
for i, source in enumerate(sources):
    plt.bar(top_10_gdp['country'], top_10_gdp[source], bottom=bottom, label=labels[i])
    bottom += top_10_gdp[source]

plt.xlabel('Country')
plt.ylabel('Percentage of Electricity Production')
plt.title('Composition of Electricity Production Sources in Top 10 GDP Countries')
plt.legend()
plt.xticks(rotation=45, ha='right')
plt.grid(axis='y')
plt.tight_layout()
plt.show()
```
