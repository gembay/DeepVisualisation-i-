```python
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
#LOAD THE DATA FROM THE SNS DATASETS
```

```python
flight_data = sns.load_dataset('flights')
```

```python
flight_data.head()
```

```python
flight_data.shape
```

```python
print(flight_data.dtypes)
print(flight_data.head())
```

```python
flight_data = flight_data.reset_index(drop=True)
```

```python
month_map = {
    'Jan': 1, 'Feb': 2, 'Mar': 3, 'Apr': 4, 'May': 5,
    'Jun': 6, 'Jul': 7, 'Aug': 8, 'Sep': 9, 'Oct': 10,
    'Nov': 11, 'Dec': 12
}
flight_data['month'] = flight_data['month'].map(month_map)
```

```python
print(flight_data.dtypes)
```

```python
flight_data = flight_data.astype(float)
```

```python
import seaborn as sns
sns.heatmap(flight_data.pivot(index='month', columns='year', values='passengers'))
```

```python
sns.heatmap(flight_data)
plt.show()
```
