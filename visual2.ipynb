```python
#WE NEED TO START VISUALISATION
```

```python
x = df[df.country == 'Kenya']
y = df[df.country == 'Tanzania']
```

```python
#WE START WITH A SIMPLE LINE GRAPH YEAR VS POPULATION
```

```python
plt.plot(x.year,x.population,label="Kenya")
plt.plot(x.year,y.population,label="Tanzania")
plt.xlabel("year")
plt.ylabel("population")

plt.legend()
plt.show()
```

```python
#SUBPLOTS >>line,histogram,bar,pie,scatter,box,heatmap
```

```python
fig ,axis = plt.subplots(2,3,figsize=(12,12))
axis[0,0].hist(x.population,bins=20)
axis[0,0].set_xlabel("frequency")
axis[0,0].set_title("histogram year vs  population")
axis[0,0].set_ylabel("population")
```

```python
axis[0,1].scatter(x.year,x.population)
axis[0,1].set_xlabel("year")
axis[0,1].set_title("scatter plot year vs  population")
axis[0,1].set_ylabel("population")
```

```python
axis[0,2].bar(x.year,x.population)
axis[0,2].set_xlabel("year")
axis[0,2].set_ylabel("population")
axis[0,2].set_title("bar graph year vs  population")
```

```python
axis[1,0].plot(x.year,x.population)
axis[1,0].set_xlabel("year")
axis[1,0].set_ylabel("population")
axis[1,0].set_title("line graph year vs  population")
```

```python
axis[1,1].boxplot(x.population)
axis[1,1].set_xlabel("year")
axis[1,1].set_ylabel("population")
axis[1,1].set_title("Box plot  population")
```

```python
labels = y.year
axis[1,2].pie(x.population,labels=labels)

axis[1,2].set_title("pie chart yr vs  population")
plt.tight_layout()
plt.show()
```

```python
sns.heatmap(x.population)
plt.show()
```
