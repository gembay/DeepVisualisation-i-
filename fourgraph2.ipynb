```python

import numpy as np

# Simulate a sample dataset
np.random.seed(42)
n = 100
data = {
    'country': [f'Country_{i}' for i in range(n)],
    'continent': np.random.choice(['Africa', 'Asia', 'Europe', 'Americas', 'Oceania'], n),
    'gdp': np.random.lognormal(mean=10, sigma=1, size=n) * 1e9,  # GDP in billions
    'life_expectancy': np.random.normal(70, 10, n).clip(50, 90),
    'co2_emissions': np.random.lognormal(mean=5, sigma=1, size=n),  # CO2 emissions (metric tons per capita)
    'electricty_production_coal_pct': np.random.uniform(0, 40, n),
    'electricty_production_hydroelectric_pct': np.random.uniform(0, 30, n),
    'electricty_production_nuclear_pct': np.random.uniform(0, 20, n),
    'electricty_production_gas_pct': np.random.uniform(0, 25, n),
    'electricty_production_oil_pct': np.random.uniform(0, 10, n),
    'electricty_production_renewable_pct': np.random.uniform(0, 30, n),
}
df = pd.DataFrame(data)
df['gdp_per_capita'] = df['gdp'] / 1e6  # Simplified GDP per capita (assuming population ~1M)

# 1. Relationship: Scatter plot of GDP per capita vs. Life Expectancy
plt.figure(figsize=(10, 6))
plt.scatter(df['gdp_per_capita'], df['life_expectancy'], alpha=0.5, color='blue')
plt.xlabel('GDP per Capita (USD)')
plt.ylabel('Life Expectancy (Years)')
plt.title('Relationship between GDP per Capita and Life Expectancy')
plt.xscale('log')  # Log scale for GDP per capita due to wide range
plt.grid(True)
plt.show()



```

```python
# 2. Comparison: Bar plot of CO2 emissions per capita across continents
co2_by_continent = df.groupby('continent')['co2_emissions'].mean()
plt.figure(figsize=(10, 6))
co2_by_continent.plot(kind='bar', color='green')
plt.xlabel('Continent')
plt.ylabel('CO2 Emissions per Capita (Metric Tons)')
plt.title('CO2 Emissions per Capita by Continent')
plt.xticks(rotation=45)
plt.grid(axis='y')
plt.show()
```

```python
# 3. Distribution: Histogram of Life Expectancy
plt.figure(figsize=(10, 6))
plt.hist(df['life_expectancy'], bins=20, color='purple', edgecolor='black')
plt.xlabel('Life Expectancy (Years)')
plt.ylabel('Frequency')
plt.title('Distribution of Life Expectancy Across Countries')
plt.grid(True)
plt.show()
```

```python

# 4. Composition: Stacked bar plot of electricity production sources for top 10 GDP countries
top_10_gdp = df.nlargest(10, 'gdp')
sources = [
    'electricty_production_coal_pct', 'electricty_production_hydroelectric_pct',
    'electricty_production_nuclear_pct', 'electricty_production_gas_pct',
    'electricty_production_oil_pct', 'electricty_production_renewable_pct'
]
labels = ['Coal', 'Hydroelectric', 'Nuclear', 'Gas', 'Oil', 'Renewable']

plt.figure(figsize=(12, 8))
bottom = np.zeros(len(top_10_gdp))
for i, source in enumerate(sources):
    plt.bar(top_10_gdp['country'], top_10_gdp[source], bottom=bottom, label=labels[i])
    bottom += top_10_gdp[source]

plt.xlabel('Country')
plt.ylabel('Percentage of Electricity Production')
plt.title('Composition of Electricity Production Sources in Top 10 GDP Countries')
plt.legend()
plt.xticks(rotation=45, ha='right')
plt.grid(axis='y')
plt.tight_layout()
plt.show()
```
